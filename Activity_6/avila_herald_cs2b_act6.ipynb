### Libraries and Functions

```python
# Import required libraries
import random
import matplotlib.pyplot as plt
from collections import Counter
import time
import sys
```

```python
dice_faces = ["⚀", "⚁", "⚂", "⚃", "⚄", "⚅"]
```

```python
def simulate_biased_dice_rolls(num_dice, weights):

    # Validate inputs
    if num_dice <= 0:
        raise ValueError("num_dice must be a positive integer")
    if len(weights) != 6:
        raise ValueError("weights must contain exactly 6 values")
    if not abs(sum(weights) - 1.0) < 0.0001:
        raise ValueError("weights must sum to 1.0")
    
    # Simulate dice rolls using random.choices with weighted probabilities
    dice_population = [1, 2, 3, 4, 5, 6]
    rolls = random.choices(dice_population, weights=weights, k=num_dice)
    
    return rolls
```

```python
def animate_dice_rolling(rolls):

    from IPython.display import clear_output
    
    print("Rolling", end="")
    sys.stdout.flush()
    
    # Animate dots
    for _ in range(3):
        time.sleep(0.3)
        print(" .", end="")
        sys.stdout.flush()
    
    print()  # New line after dots
    
    # Display the rolls 20 per line with animated updates
    dice_per_line = 20
    num_frames = 10  # Number of animation frames
    
    # Animate: show random dice changing to create rolling effect
    for frame in range(num_frames):
        clear_output(wait=True)
        print("Rolling . . .")
        
        # Show random dice for animation effect (except last frame)
        if frame < num_frames - 1:
            temp_rolls = [random.randint(1, 6) for _ in range(len(rolls))]
        else:
            temp_rolls = rolls  # Final frame shows actual results
        
        # Display all dice
        for i in range(0, len(temp_rolls), dice_per_line):
            line_rolls = temp_rolls[i:i+dice_per_line]
            dice_symbols = ''.join([dice_faces[roll-1] for roll in line_rolls])
            print(dice_symbols)
        
        time.sleep(0.15)  # Brief pause between frames
    
    # Clear the animation output after it's done
    clear_output(wait=True)
    
    # Show final rolling result one more time (this will be the only persistent output)
    print("Rolling . . .")
    for i in range(0, len(rolls), dice_per_line):
        line_rolls = rolls[i:i+dice_per_line]
        dice_symbols = ''.join([dice_faces[roll-1] for roll in line_rolls])
        print(dice_symbols)
```

```python
def print_sorted_dice_rolls(sorted_dice):

    print("\nSorted Dice Rolls:")
    
    # Convert numbers to dice face symbols and display 20 per line
    dice_per_line = 20
    for i in range(0, len(sorted_dice), dice_per_line):
        line_rolls = sorted_dice[i:i+dice_per_line]
        dice_symbols = ''.join([dice_faces[roll-1] for roll in line_rolls])
        print(dice_symbols)
```

```python
def count_dice_face_frequency(sorted_dice):

    # Count frequencies using Counter
    frequency = Counter(sorted_dice)
    
    # Ensure all faces (1-6) are represented, even if count is 0
    frequency_dict = {i: frequency.get(i, 0) for i in range(1, 7)}
    
    return frequency_dict
```

```python
def calculate_dice_sum(sorted_dice):

    return sum(sorted_dice)
```

```python
def plot_dice_roll_distribution(sorted_dice):

    # Calculate sum for the title
    dice_sum = sum(sorted_dice)
    
    # Count frequencies to find the maximum
    frequency = Counter(sorted_dice)
    max_frequency = max(frequency.values()) if frequency else 0
    
    # Set up the plot with professional styling
    plt.figure(figsize=(8, 6))
    
    # Create histogram with custom styling and spacing between bars
    counts, bins, patches = plt.hist(sorted_dice, bins=range(1, 8), 
                                      edgecolor='black', alpha=0.7, 
                                      color='#1f77b4', align='left',
                                      rwidth=0.8)
    
    # Use numbers instead of dice symbols for x-axis
    plt.xticks(range(1, 7), [1, 2, 3, 4, 5, 6], fontsize=12)
    
    # Set y-axis range dynamically based on maximum frequency (add 10% padding)
    plt.ylim(0, max_frequency * 1.1)
    
    # Customize the plot
    plt.xlabel('Dice Face', fontsize=14, fontweight='bold')
    plt.ylabel('Frequency', fontsize=14, fontweight='bold')
    plt.title(f'Dice Roll Distribution (SUM: {dice_sum})', fontsize=16, fontweight='bold', pad=20)
    
    # Add value labels on top of each bar
    for i, count in enumerate(counts):
        plt.text(i + 1, count + 1, str(int(count)), 
                ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
```

```python
def find_most_and_least_frequent_faces(sorted_dice):
    
    # Count frequencies
    frequency = Counter(sorted_dice)
    
    # Ensure all faces are represented
    frequency_dict = {i: frequency.get(i, 0) for i in range(1, 7)}
    
    # Find most and least frequent faces
    most_frequent_face = max(frequency_dict, key=frequency_dict.get)
    least_frequent_face = min(frequency_dict, key=frequency_dict.get)
    
    # Display results (exactly as specified in requirements)
    print(f"\nMost Frequent Face: {most_frequent_face}")
    print(f"Least Frequent Face: {least_frequent_face}")
    
    return most_frequent_face, least_frequent_face
```

```python
# Configuration: Set simulation parameters
num_dice = 123  # Number of dice rolls to simulate (6 lines of 20 + 1 line of 3)
weights = [0.4, 0.2, 0.1, 0.1, 0.1, 0.1]  # Probabilities for faces 1-6

# Step 1: Simulate biased dice rolls
rolls = simulate_biased_dice_rolls(num_dice, weights)

# Step 2: Animate the rolling process (real-time visual)
animate_dice_rolling(rolls)

# Step 3: Sort the rolls and display them
sorted_dice = sorted(rolls)
print_sorted_dice_rolls(sorted_dice)

# Step 4: Count frequencies (silent - used for analysis)
frequency_dict = count_dice_face_frequency(sorted_dice)

# Step 5: Calculate sum (silent - used for statistics)
total_sum = calculate_dice_sum(sorted_dice)

# Step 6: Find and display most/least frequent faces
most_freq, least_freq = find_most_and_least_frequent_faces(sorted_dice)

# Step 7: Plot the distribution
plot_dice_roll_distribution(sorted_dice)
```
