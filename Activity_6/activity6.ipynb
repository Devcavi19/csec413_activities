```python
# Import required libraries
import random
import matplotlib.pyplot as plt
from collections import Counter
```

```python
dice_faces = ["⚀", "⚁", "⚂", "⚃", "⚄", "⚅"]
```

## Task 1: Simulate Biased Dice Rolls
This function simulates rolling a biased six-sided dice with specified weights for each face.

```python
def simulate_biased_dice_rolls(num_dice, weights):
    """
    Simulates rolling a biased six-sided dice with specified weights.
    
    Parameters:
    - num_dice (int): Number of dice rolls to simulate
    - weights (list): List of 6 floats representing probabilities for each face (must sum to 1.0)
    
    Returns:
    - list: List of integers (1-6) representing the dice roll results
    """
    # Validate inputs
    if num_dice <= 0:
        raise ValueError("num_dice must be a positive integer")
    if len(weights) != 6:
        raise ValueError("weights must contain exactly 6 values")
    if not abs(sum(weights) - 1.0) < 0.0001:  # Allow for floating point precision
        raise ValueError("weights must sum to 1.0")
    
    # Simulate dice rolls using random.choices with weighted probabilities
    dice_population = [1, 2, 3, 4, 5, 6]
    rolls = random.choices(dice_population, weights=weights, k=num_dice)
    
    return rolls
```

## Task 2: Print Sorted Dice Rolls
This function displays the sorted dice rolls in a structured format using dice face symbols.

```python
def print_sorted_dice_rolls(sorted_dice):
    """
    Displays sorted dice rolls in a structured format using dice face symbols.
    Prints 20 dice faces per line for better readability.
    
    Parameters:
    - sorted_dice (list): List of integers (1-6) representing sorted dice rolls
    
    Returns:
    - None (prints output)
    """
    print("\n" + "="*60)
    print("SORTED DICE ROLLS")
    print("="*60)
    
    # Convert numbers to dice face symbols and display 20 per line
    dice_per_line = 20
    for i in range(0, len(sorted_dice), dice_per_line):
        line_rolls = sorted_dice[i:i+dice_per_line]
        dice_symbols = ''.join([dice_faces[roll-1] for roll in line_rolls])
        print(dice_symbols)
    
    print("="*60 + "\n")
```

## Task 3: Count Frequency of Each Dice Face
This function counts how many times each dice face appears in the rolls.

```python
def count_dice_face_frequency(sorted_dice):
    """
    Counts the frequency of each dice face in the sorted rolls.
    
    Parameters:
    - sorted_dice (list): List of integers (1-6) representing sorted dice rolls
    
    Returns:
    - dict: Dictionary with dice faces as keys and their frequencies as values
    """
    # Count frequencies using Counter
    frequency = Counter(sorted_dice)
    
    # Ensure all faces (1-6) are represented, even if count is 0
    frequency_dict = {i: frequency.get(i, 0) for i in range(1, 7)}
    
    # Display the frequency table
    print("\n" + "="*60)
    print("DICE FACE FREQUENCY")
    print("="*60)
    print(f"{'Dice Face':<15} {'Symbol':<10} {'Frequency':<15} {'Percentage':<10}")
    print("-"*60)
    
    total_rolls = len(sorted_dice)
    for face in range(1, 7):
        count = frequency_dict[face]
        percentage = (count / total_rolls * 100) if total_rolls > 0 else 0
        print(f"{'Face ' + str(face):<15} {dice_faces[face-1]:<10} {count:<15} {percentage:.2f}%")
    
    print("="*60 + "\n")
    
    return frequency_dict
```

## Task 4: Calculate Sum of Dice Rolls
This function computes the total sum of all dice rolls.

```python
def calculate_dice_sum(sorted_dice):
    """
    Computes the total sum of all dice rolls.
    
    Parameters:
    - sorted_dice (list): List of integers (1-6) representing sorted dice rolls
    
    Returns:
    - int: Total sum of all dice rolls
    """
    total_sum = sum(sorted_dice)
    
    # Calculate expected value and average
    num_rolls = len(sorted_dice)
    average = total_sum / num_rolls if num_rolls > 0 else 0
    
    print("\n" + "="*60)
    print("DICE ROLL STATISTICS")
    print("="*60)
    print(f"Total Number of Rolls: {num_rolls}")
    print(f"Sum of All Rolls:      {total_sum}")
    print(f"Average Roll Value:    {average:.2f}")
    print("="*60 + "\n")
    
    return total_sum
```

## Task 5: Plot Dice Roll Distribution
This function creates a histogram to visualize the distribution of dice rolls.

```python
def plot_dice_roll_distribution(sorted_dice):
    """
    Creates a histogram to visualize the distribution of dice rolls.
    
    Parameters:
    - sorted_dice (list): List of integers (1-6) representing sorted dice rolls
    
    Returns:
    - None (displays plot)
    """
    # Set up the plot with professional styling
    plt.figure(figsize=(12, 6))
    
    # Create histogram with custom styling
    counts, bins, patches = plt.hist(sorted_dice, bins=range(1, 8), 
                                      edgecolor='black', alpha=0.7, 
                                      color='steelblue', align='left')
    
    # Add dice face symbols as x-tick labels
    plt.xticks(range(1, 7), [dice_faces[i] for i in range(6)], fontsize=20)
    
    # Customize the plot
    plt.xlabel('Dice Face', fontsize=14, fontweight='bold')
    plt.ylabel('Frequency', fontsize=14, fontweight='bold')
    plt.title('Distribution of Biased Dice Rolls', fontsize=16, fontweight='bold', pad=20)
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Add value labels on top of each bar
    for i, count in enumerate(counts):
        plt.text(i + 1, count + max(counts)*0.01, str(int(count)), 
                ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    # Add statistics text box
    total_rolls = len(sorted_dice)
    average = sum(sorted_dice) / total_rolls if total_rolls > 0 else 0
    stats_text = f'Total Rolls: {total_rolls}\nAverage: {average:.2f}'
    plt.text(0.98, 0.97, stats_text, transform=plt.gca().transAxes,
            fontsize=10, verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
```

## Task 6: Identify Most and Least Frequent Dice Faces
This function determines which dice face appears most and least frequently.

```python
def find_most_and_least_frequent_faces(sorted_dice):
    """
    Identifies the dice face that occurs most frequently and least frequently.
    
    Parameters:
    - sorted_dice (list): List of integers (1-6) representing sorted dice rolls
    
    Returns:
    - tuple: (most_frequent_face, least_frequent_face)
    """
    # Count frequencies
    frequency = Counter(sorted_dice)
    
    # Ensure all faces are represented
    frequency_dict = {i: frequency.get(i, 0) for i in range(1, 7)}
    
    # Find most and least frequent faces
    most_frequent_face = max(frequency_dict, key=frequency_dict.get)
    least_frequent_face = min(frequency_dict, key=frequency_dict.get)
    
    # Display results
    print("\n" + "="*60)
    print("MOST AND LEAST FREQUENT FACES")
    print("="*60)
    print(f"Most Frequent Face:  {most_frequent_face} {dice_faces[most_frequent_face-1]} (appeared {frequency_dict[most_frequent_face]} times)")
    print(f"Least Frequent Face: {least_frequent_face} {dice_faces[least_frequent_face-1]} (appeared {frequency_dict[least_frequent_face]} times)")
    print("="*60 + "\n")
    
    return most_frequent_face, least_frequent_face
```
