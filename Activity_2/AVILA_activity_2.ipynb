**Task Description:**

Create a Python program that simulates a computer's attempts to guess a secret number chosen randomly between 1 and 100,000. The program offers two modes of guessing:

**Random Guess Mode:** In this mode (mode 1), the computer makes random guesses within the given range and adjusts its guesses based on whether they are too high or too low compared to the secret number.

**Binary Search Mode:** In this mode (mode 2), the computer employs a binary search algorithm to systematically narrow down its guesses by always choosing the midpoint of the current search range.

The program keeps track of the number of attempts made by the computer and plots the progress of the guesses over the attempts using Matplotlib. It updates the plot in real-time as the computer makes guesses.

**The main components and features of the code include:**
Randomly generating a secret number between 1 and 100,000.
A guess() function that takes the mode as input (1 for random guess, 2 for binary search).
Tracking the computer's attempts and guesses and storing them in lists for plotting.
Handling messages to inform whether the computer's guess was too high or too low.
Continuously updating and displaying a plot of the computer's guesses over attempts.
Allowing the user to choose between the two modes (random guess or binary search) and passing the selected mode to the guess() function.

**To use the program:**
Run the code.\
Choose a mode (1 for random guess or 2 for binary search) when prompted.\
Watch the computer make guesses and see the plot of its progress in real-time.\
The program will terminate when the computer successfully guesses the secret number, and you'll see the number of attempts it took.\
Note: The program uses Matplotlib for plotting and clears the output and updates the plot after each guess, providing a dynamic visualization of the guessing process.\
It uses a time delay of 0.2 seconds between updates to make the process visually understandable. The binary search mode is expected to find the secret number more quickly compared to the random guess mode.


```python
import random
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
def guess(mode):
    """
    Main guessing function that implements both random and binary search algorithms
    
    Parameters:
    mode (int): 1 for random guess mode, 2 for binary search mode
    """
    # Generate random secret number between 1 and 100,000
    secret_number = random.randint(1, 100000)
    
    # Initialize tracking variables
    attempts = []
    guesses = []
    attempt_count = 0
    
    # Initialize range boundaries for both modes
    low = 1
    high = 100000
    
    # Determine mode name for display
    mode_name = "Random Guess" if mode == 1 else "Binary Search"
    
    # Main guessing loop
    while True:
        attempt_count += 1
        
        # Generate guess based on selected mode
        if mode == 1:  # Random Guess Mode
            # Make random guess within current valid range
            current_guess = random.randint(low, high)
            
        elif mode == 2:  # Binary Search Mode
            # Calculate midpoint for binary search
            current_guess = (low + high) // 2
        
        # Store attempt data for plotting
        attempts.append(attempt_count)
        guesses.append(current_guess)
        
        # Check if guess is correct
        if current_guess == secret_number:
            # Create final success plot - keep blue color
            plt.figure(figsize=(10, 6))
            plt.plot(attempts, guesses, 'b-o', markersize=6, linewidth=2)
            plt.xlabel('Attempt Number')
            plt.ylabel('Guess Value')
            # Updated title format for correct guess
            plt.title(f"Attempt {attempt_count} - Computer's {mode_name} {current_guess}")
            plt.ylim(0, 100000)
            plt.show()
            break
            
        # Provide feedback and adjust search range
        elif current_guess < secret_number:
            feedback = "is too low"
            low = current_guess + 1  # Narrow the lower bound
            
        else:  # current_guess > secret_number
            feedback = "is too high"
            high = current_guess - 1  # Narrow the upper bound
        
        # Create and update real-time plot with simplified format
        plt.figure(figsize=(10, 6))
        plt.plot(attempts, guesses, 'b-o', markersize=5, linewidth=2)
        plt.xlabel('Attempt Number')
        plt.ylabel('Guess Value')
        # Updated title format with guess value and feedback
        plt.title(f"Attempt {attempt_count} - Computer's {mode_name} {current_guess} - {feedback}")
        plt.ylim(0, 100000)
        plt.show()
        
        # Add delay and clear output for animation effect
        time.sleep(0.2)
        clear_output(wait=True)
```

## Simulation Results in Random Guess Mode

Random Guess is option 1 in number guessing simulation. The algorithm located the secret number **39480** in exactly **24 attempts**.

```python
if __name__ == "__main__":
    """
    Main program execution with user interface
    """
    print("Computer Modes")
    print("1. Random Guess")
    print("2. Binary Search")
    
    # Input validation loop
    while True:
        try:
            mode = int(input("Enter mode (1 or 2): "))
            
            if mode == 1:
                break
            elif mode == 2:
                break
            else:
                print("❌ Invalid choice! Please enter 1 or 2.")
                
        except ValueError:
            print("❌ Invalid input! Please enter a number (1 or 2).")
    
    # Run the guess function with selected mode
    guess(mode)
```
